```python
#LIBRERIAS 
import serial
import csv
from datetime import datetime
```

```python
#PARÁMETROS
MAXTIME = 10
PUERTO_SERIE = 'COM9'
VELOCITY = 57600
```

PARTE 1 PARA OBTENER DATOS:
Adquisicion de datos de la mano robótica....


```python

# Configura el puerto serie - ajusta el puerto y la velocidad (baudrate) según corresponda
puerto_serie = serial.Serial(PUERTO_SERIE, VELOCITY)  # Reemplaza 'COMX' con el nombre de tu puerto serie segun al que se conecte tu arduino
                                            #9600 no sé si está bien. Se calculaba en funcion del teorema de Nyquist.
                                            #explicar por qué se cambia la velocidad del puerto serie


# Abre un archivo CSV para escribir los datos EMG
#CATEGORIA 0-->REPOSO
#CATEGORIA 1-->TORSION HACIA ABAJO
#CATEGORIA 2-->TORSION HACIA ARRIBA 
#CATEGORIA 3-->APRETAR, PUÑO CERRADO
#CATEGORIA 4--
categoria = 3
timestamp = datetime.now().timestamp()
print(timestamp)

name = ''.join(["datos_emg_", str(categoria),'_',str(timestamp),'.csv'])
archivo_csv = open(name, 'w', newline='') #se crea desde cero, o añadir la 'a', para añadirtelo si ya está creado
csv_writer = csv.writer(archivo_csv)

try:
  start = datetime.now()
  cont = 0 
  while True:
        current = datetime.now()
        prueba = (current - start).seconds

        if(prueba > MAXTIME):
           break

        # Lee los datos desde el puerto serie
        linea_serial = puerto_serie.readline().decode().strip()  # Lee una línea y decodifica los bytes a una cadena(string)

        # Suponiendo que los datos están separados por comas en el formato "dato1, dato2, dato3, ..."
        #datos_emg = linea_serial.split(',')  # Separa los datos por comas #comprobar que se quieran dividir los datos de esta forma

        # Escribe los datos en el archivo CSV
        archivo_csv.write(linea_serial) #escribimos directamente con el archivo creado que es archivo_csv
        archivo_csv.write("\n") 

        # Imprime los datos recibidos en la consola (opcional)
        print(linea_serial)
        cont += 1

except KeyboardInterrupt:
    print("Interrupción del usuario. Cerrando el programa.")

finally:
    print("captura finalizada")
    print("Muestras", cont )
    # Cierra el puerto serie y el archivo CSV al finalizar
    puerto_serie.close()
    archivo_csv.close()
```
